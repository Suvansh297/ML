```python
import numpy as np
import pandas as pd
import requests
```

```python
Username = input()
str = 'https://codeforces.com/api/user.status?handle=' + Username + '&from=1&count=100000'
str
```

```python
req = requests.get(str)
```

```python
df = pd.DataFrame(req.json()['result'])
```

```python
list0 = []
col = df.shape[0]
for i in range (col):
    
    list1 = df['problem'][i]
    if 'rating' in list1.keys():
        list0.append(list1['rating'])
    else:
        list0.append(None)
        

    
    
```

```python
list0
```

```python

df.insert(6, "Rating", list0, True)
```
